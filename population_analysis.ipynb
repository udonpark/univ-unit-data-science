```python

w = 0.2 #a width of the graph
bar1 = np.arange(len(data_q2))
bar2 = [w + i for i in bar1]
bar3 = [w + i for i in bar2]
q2_country=data_q2['country']
q2_population=data_q2['population']
q2_total=data_q2['total_vaccinations']
q2_fully=data_q2['people_fully_vaccinated']

plt.bar(bar1, q2_population, w, label = 'Population in Thousands')
plt.bar(bar2, q2_total, w, label = 'Total Vaccinations')
plt.bar(bar3, q2_fully, w, label = 'People fully vaccinated')

plt.xticks(bar1, q2_country, rotation = 80)
plt.xlabel('Countries')
plt.ylabel('Population')

plt.legend()
plt.title('Data for each country')
plt.show()

#Upon writing this code, I have looked on this video to get the general view. https://www.youtube.com/watch?v=ZjQCPMO7LBE
```

Now the printing side by side bar graph is done, however I see some challenges here as specified in the instruction pdf.

First, bars for indonesia is pulling up the maximum value of the graph too much. It is too dominant that it is hard to visually see the differences in other countries.

This can be solved by setting max and min values for the y value.
For the practical codes, I have refered to:
https://stackoverflow.com/questions/3777861/setting-y-axis-limit-in-matplotlib

```python
w = 0.2 #a width of the graph
bar1 = np.arange(len(data_q2))
bar2 = [w + i for i in bar1]
bar3 = [w + i for i in bar2]
q2_country=data_q2['country']
q2_population=data_q2['population']
q2_total=data_q2['total_vaccinations']
q2_fully=data_q2['people_fully_vaccinated']

plt.bar(bar1, q2_population, w, label = 'Population in Thousands')
plt.bar(bar2, q2_total, w, label = 'Total Vaccinations')
axes = plt.gca()
axes.set_ylim([0,380000])
plt.bar(bar3, q2_fully, w, label = 'People fully vaccinated')
axes = plt.gca()
axes.set_ylim([0,150000])

plt.xticks(bar1, q2_country, rotation = 80)
plt.xlabel('Countries')
plt.ylabel('Population')

plt.legend()
plt.title('Data for each country')
plt.show()
```

Re-scaling has been done and the first challenge is solved. However, the downside of this is that It is hard to see how much value Indonesia and Singapore has, without actually seeing the raw data. Still, it is way better to be able to compare ignoring the dominating factors.

Secondly, this visualization is not perfectly intuitive. That is because there are 3 types of data represented here, and the unit for 'population'(Thousands) is different from that of 'total_vaccinations' and 'people_fully_vaccinated'(One person).


We can solve this second problem by adding a label on the right as well as a different scale. I will use subplots() to make this possible.

```python
w = 0.2 #a width of the graph
bar1 = np.arange(len(data_q2)) #set an array of range(len)
bar2 = [w + i for i in bar1]
bar3 = [w + i for i in bar2]
q2_country=data_q2['country']
q2_population=data_q2['population']
q2_total=data_q2['total_vaccinations']
q2_fully=data_q2['people_fully_vaccinated'] #set variables so that it is easy to handle

fig, ax1 = plt.subplots()  #prepare a subplot

ax1.bar(bar1, q2_population, w, label = 'Population in Thousands', color='b')
ax1.set_xlabel('Countries')
ax1.set_ylabel('Population in Thousands')
plt.xticks(bar1, q2_country, rotation = 80)
plt.legend(loc='upper left')

ax2 = ax1.twinx()  #make a twin subplot, sharing the same x axis
ax2.bar(bar2, q2_total, w, label = 'Total Vaccinations', color='g')
ax2.set_ylabel('Number of people')
axes = plt.gca()
axes.set_ylim([0,380000])
ax2.bar(bar3, q2_fully, w, label = 'People fully vaccinated', color='orange')
axes = plt.gca()
axes.set_ylim([0,150000])


plt.xticks(bar1, q2_country, rotation = 80)

plt.legend(loc='upper right') #make sure legend is in different location
plt.title('Data for each country')
plt.show()

#Reference to : https://matplotlib.org/stable/gallery/subplots_axes_and_figures/two_scales.html
```

### A lot better! 3 Bars are neatly seperated into 2 parts, right and left. Scales and legends are easy to see after overcoming 2 challenges.

Lastly, I would note down some possibility why this data could be misleading.

But because of that, one reader might misinterrupt after seeing the green bar(total vaccinations) slightly less than blue bar(population) in Cambodia and consider that "Oh, almost all population of cambodia has already done vaccinations".
Which is not true; you can say that one in a thousand of people in Cambodia has done vaccinations instead.

- Because of having data of only 4 countries out of 11 who has started vaccination process, It is not really easy to find correlations or relations to perform an actual analysis on these data. Maybe the good idea is to wait until the data is updated to collect and perform the graphing again.
